```python
# filling the missing values unsing Median 
#df['Review_Rating'].fillna(df['Review_Rating'].median(), inplace=True)
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# chesking the null values 
df.isnull().sum()
```

```python
# converting columns into lowercase and replace spaces with under score'_'
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# Feature Engineering 
# Create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels) # qcuts split the ages into 4 categories
```

```python
df[['age', 'age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
#  create column purchase _frequence_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequently_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
