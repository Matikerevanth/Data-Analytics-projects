```python
# importing libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# loading the dataset
dataset = pd.read_csv(r"C:\Users\jrbil\OneDrive\Desktop\Data science\UDEMY\project\orders.csv")
```

```python
dataset.head(20)
```

```python
# finding the missing values
dataset.isnull().sum()
```

```python
# finding the unique values in Ship Mode
dataset['Ship Mode'].unique()
```

```python
# replacing the unknown values to NA
dataset.replace(['Not Available', 'nan',  'unknown' , '--'],pd.NA, inplace = True)
```

```python
dataset.head(20)
```

```python
print(dataset.columns)
```

```python
#dataset.columns.str.lower()
dataset.columns = dataset.columns.str.replace(' ', '_')
```

```python
print(dataset.columns)
```

```python
# creating the new column discount
dataset['discount']=dataset['List_Price']*dataset['Discount_Percent']/100
```

```python
dataset
```

```python
# creating new column of saling price
dataset['sale_price']=dataset['List_Price']-dataset['discount']
```

```python
dataset
```

```python
dataset['profit'] = dataset['sale_price']-dataset['cost_price']
```

```python
dataset
```

```python
dataset.dtypes
```

```python
dataset['Order_Date']=pd.to_datetime(dataset['Order_Date'],format="%Y-%m-%d")
```

```python
dataset.dtypes
```

```python
# droping the unnessary columns
dataset.drop(columns=['List_Price', 'cost_price', 'Discount_Percent'], inplace=True)
```
