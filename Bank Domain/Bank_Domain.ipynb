```python
for col in categorical_cols:
    plt.figure(figsize=(10,6))
    value_counts = bank_df[col].value_counts()
    plt.bar(value_counts.index, value_counts.values)
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.title(f'{col} - Bar Chart')
    plt.xticks(rotation=45)
    plt.show()
```

```python
categorical_cols=bank_df.select_dtypes(include='object').columns
numerical_cols=bank_df.select_dtypes(exclude='object').columns
n=len(categorical_cols)-1
plt.figure(figsize=(28,14))
for i in range(n):
    labels=bank_df[categorical_cols[i+1]].value_counts().keys()
    values=bank_df[categorical_cols[i+1]].value_counts().values
    plt.subplot(3,3,i+1)
    plt.bar(labels,values)
    plt.xlabel('labels')
    plt.ylabel('count')
    plt.title(f'{categorical_cols[i]}_Bar_chart')
    plt.show()
```

```python
import math

n = len(categorical_cols)
cols = 3                  # number of columns in subplot grid
rows = math.ceil(n / cols)  # automatically calculate rows

plt.figure(figsize=(5*cols, 5*rows))

for i in range(n):
    plt.subplot(rows, cols, i+1)
    value_counts = bank_df[categorical_cols[i]].value_counts()
    plt.pie(value_counts.values, labels=value_counts.index, autopct='%0.2f%%')
    plt.title(categorical_cols[i])

plt.tight_layout()
plt.show()
```
